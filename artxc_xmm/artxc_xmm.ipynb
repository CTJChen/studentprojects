```python
import astropy.coordinates as cd
import astropy.units as u
from astropy.io import fits
from astropy.table import Table
import numpy as np
```

```python
# This is from one of the sources in the google drive link
# we are making a "SkyCoord ojbect" to be used later
coord_art = cd.SkyCoord(264.28938, 66.01817, unit=(u.degree,u.degree))
```

##### I have downloaded the PPS product from the XMM website, it's usually a packaged file with a .tar extension. 
You should be able to extract its content by typing "tar -xvf {filename}.tar"  (substite {filename} with the real filename.) 
Once you extracted the files, we can see if there's a known XMM source near the position defined by coord_art


```python

# Let's load the source list into memory
hdu = fits.open('/Users/cchen24/Downloads/xmm/0202100301/pps/P0202100301EPX000OBSMLI0000.FTZ')
```

```python
# Fits or FTZ file can have multiple extensions, we can see what the content is by the following command:
hdu.info()
```

```python
# converting to astropy table for readability
xmmlist = Table(hdu[1].data)
xmmlist
```

```python
# we can make a skycoord object for all the XMM sources in this list
coord_xmm = cd.SkyCoord(xmmlist['RA'], xmmlist['DEC'], unit=(u.degree,u.degree))
```

```python
# then we can match the sky positions between coord_art and coord_xmm.
# Here I set the unit of the result to "arcsec"
distance_arcsec = coord_art.separation(coord_xmm).arcsec
```

```python
distance_arcsec
```

```python
# See if the closest XMM object is reasonable close to coord_nus - the positional uncertainty for ART-XC is about 30"
print(np.min(distance_arcsec))
```

```python
# if we want to access the row in the xmmlist table that's closest to coord_art, we can do the following:
xmmsrc = xmmlist[np.argmin(distance_arcsec)]
```

```python
# so we know it's source number 1 (SRC_NUM=1), 
xmmsrc
```

```python
import os
flist = os.listdir('/Users/cchen24/Downloads/xmm/0202100301/pps/')
# I'm interested in the source spectrum of the EPIC-PN instrument (the most sensitive one on XMM):
# and we only want SRC_NUM=1, so the string to search for is "SRSPEC0001"
speclist = [i for i in flist if 'SRSPEC0001' in i]
print(speclist)
```

On my machine, the file that we want to load in xspec is "/Users/cchen24/Downloads/xmm/0202100301/pps/P0202100301PNS003SRSPEC0001.FTZ" 
The two files with M1 or M2 in their file names are from different instruments. They are usually similar to PN spectra but you can load them in xspec if you want.
When you load them in XSPEC, you might see something like 
"Error: cannot read response file epn_e1_ff20_sdY9.rmf"

You can search the file name on google, which should take you to an XMM website where you can download it and put it into the same directory as your spectral files.
If you have the .rmf file already, the error message shouldn't show, but it can also be loaded by typing "resp epn_e1_ff20_sdY9.rmf" under the XSPEC prompt.

```python
# Note the auxillary files are also not linked automatically to the spectral files, but they should be under the same PPS directory
arflist = [i for i in flist if 'SRCARF0001' in i]
print(arflist)
# once you've located the correct ARF file, you can link it to the spectral file in xspec by typing "arf P0202100301PNS003SRCARF0001.FTZ"
```
